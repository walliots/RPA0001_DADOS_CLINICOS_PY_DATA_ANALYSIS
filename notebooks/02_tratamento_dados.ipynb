Aplica as regras do enunciado sobre o dataset integrado (`datasets/processed/dados_integrados.csv`):
- remove linhas duplicadas
- preenche valores ausentes em colunas categóricas com a moda
- preenche valores ausentes em colunas numéricas com a mediana

```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent / "src"))

import pandas as pd
from tratamento import ENTRADA, preencher_ausentes, remover_duplicatas

df = pd.read_csv(ENTRADA, sep=";", encoding="utf-8-sig")
df.shape
```

## Diagnóstico antes do tratamento

```python
print("Linhas duplicadas:", df.duplicated().sum())
print()
print("Valores ausentes por coluna:")
print(df.isna().sum()[df.isna().sum() > 0])
```

## Aplicando o tratamento

`classe_trabalho` (categórica) → preenchida com a moda.
`qtde_filhos` (numérica) → preenchida com a mediana.
Linhas duplicadas (herdadas do join) são removidas.

```python
df_tratado = remover_duplicatas(df)
df_tratado = preencher_ausentes(df_tratado)

print("Linhas antes:", len(df), "| depois:", len(df_tratado))
print("Nulos restantes:", df_tratado.isna().sum().sum())
df_tratado.head()
```

## Salvando o dataset tratado
